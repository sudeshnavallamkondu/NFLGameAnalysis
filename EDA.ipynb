```python
import pandas as pd

data = pd.read_csv('data/processed/averages/last_10_games.csv')
```

```python
data.head()
```

```python
#drop the data where last_10_games_home_score is NaN or 0
data = data.dropna(subset=['last_10_games_home_score'])
data = data[data['last_10_games_home_score'] != 0]

data.head()
```

```python
#Descriptive statistics
data.describe()
```

```python
#Correlation Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Is there a home_field advantage? Check how the results are distributed on a pie chart
plt.figure(figsize=(12, 10))
data['result'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Results Distribution')
plt.show()
```

```python
data['result'] = data['result'].map({'home_win': 1, 'tie': 0, 'away_win': -1})
```

```python
#Visualize the top 10 correlating features to result

#map the result to 0 , 1, and -1
plt.figure(figsize=(12, 10))
data.corr()['result'].sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 correlating features to result')
plt.show()
```

```python
#Distribution of home_score and away_score
plt.figure(figsize=(12, 10))
sns.kdeplot(data['last_10_games_home_score'], label='home_score')
sns.kdeplot(data['last_10_games_away_score'], label='away_score')
plt.title('Distribution of home_score and away_score')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='last_10_games_home_score', y='result', data=data)
plt.title('Home Score vs. Game Result')
plt.show()

sns.scatterplot(x='last_10_games_away_score', y='result', data=data)
plt.title('Away Score vs. Game Result')
plt.show()
```

```python
sns.boxplot(x='result', y='last_10_games_home_score', data=data)
plt.title('Home Score Distribution by Game Outcome')
plt.show()


sns.boxplot(x='result', y='last_10_games_away_score', data=data)
plt.title('Away Score Distribution by Game Outcome')
plt.show()
```
